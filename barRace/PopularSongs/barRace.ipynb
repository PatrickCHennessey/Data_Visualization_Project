```python
def binningFunc(col):
    col = float(col)
    if col > 49:
        return 1
    elif col > 25:
        return 2
    elif col > 10:
        return 3
    elif col > 5:
        return 4
    elif col < 1:
        return 5
    else:
        return 6
```

```python
b_df5 = b_df5.assign(bins = lambda x: x["Weekly.rank"].apply(binningFunc))
```

```python
b_df5.head()
```

```python
def popularityPoints(column):
    if column == 1:
        return (abs(100 - column)+10)
    else:
        return abs(100 - column)
```

```python
b_df5 = b_df5.assign(popPoints = lambda x: x["Weekly.rank"].apply(popularityPoints))
```

```python
b_df5["POP"] = b_df5["bins"] * b_df5["popPoints"]
```

```python
b_df5.to_csv("popularity.csv")
```
