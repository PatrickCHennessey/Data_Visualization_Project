```python
a = pd.read_csv(hot_100)
b = pd.read_csv(song_attribute)
merged = a.merge(b, on='Name')
merged_df = a.merge(b, left_on=["Artists", "Name"], right_on=['Artist', "Name"]).copy()
```

```python
# Drop writing credits, lyrics, explicit, unnamed:0_x, and index",
merged_df = merged_df.drop(['Writing.Credits',
                            'Explicit',
                            'Lyrics',
                            'Date',
                            'Features',
                            'Unnamed: 0_x'], axis=1).copy()
```

```python
merged_df.columns
```

```python
new_merged_df = pd.DataFrame({"Artists":(merged_df["Artists"]),"Name":(merged_df["Name"])
                              ,"Peak Position":(merged_df["Peak.position"]), "Week":(merged_df["Week"]), 
                              "Genre":(merged_df["Genre"]),"Danceability":(merged_df["Danceability"]),
                              "Duration":(merged_df["Duration"]), "Energy":(merged_df["Energy"]),
                              "Liveness":(merged_df["Liveness"]), "Loudness":(merged_df["Loudness"]),
                              "Mode":(merged_df["Mode"]),"Popularity":(merged_df["Popularity"]),
                              "Tempo":(merged_df["Tempo"]), "Valence":(merged_df["Valence"])})
```
