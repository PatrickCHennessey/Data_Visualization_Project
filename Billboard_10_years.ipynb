```python
a = pd.read_csv(hot_100)
b = pd.read_csv(song_attribute)

merged = a.merge(b, on='Name')

merged_df = a.merge(b, left_on=["Artists", "Name"], right_on=['Artist', "Name"]).copy()
```

```python
merged_df
```

```python
# Drop writing credits, lyrics, explicit, unnamed:0_x, and index
merged_df = merged_df.drop(['Writing.Credits',
                            'Explicit', 
                            'Lyrics', 
                            'Date',
                            'Features',
                            'Unnamed: 0_x'], axis=1).copy()
```
